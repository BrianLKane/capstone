```python
mike_trout_2019 = batters_2019_daily_standard[batters_2019_daily_standard['Name'] == 'Mike Trout']
```

```python
mike_trout_2019.head()
```

```python
mike_trout_2019['Date'] = pd.to_datetime(mike_trout_2019['Date'])
```

```python
mike_trout_2019.dtypes
```

```python
mike_trout_2019.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(16, 8))

plt.plot(mike_trout_2019['AVG'])
plt.title('Mike Trout - 2019 Daily Batting Average', fontsize=18);
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
trout_model = ARIMA(endog = mike_trout_2019['AVG'].diff(4).dropna(),
              order = (1,0,0))
```

```python
arima = trout_model.fit()
preds = arima.predict()
```

```python
preds.head()
```

```python
plt.figure(figsize=(16,8))
plt.plot(mike_trout_2019['AVG'].diff(4), color = 'blue')
plt.plot(preds, color = 'orange')
plt.title(label = 'Mike Trout Daily AVG with ARIMA Predictions', fontsize=18)
plt.show();
```

```python
custom_df = pd.DataFrame(columns=batters_2019_daily_standard.columns)
```

```python
for player in batters_2019_daily_standard.playerId.unique():
    player_df = batters_2019_daily_standard[batters_2019_daily_standard['playerId'] == player]
    player_df['Date'] = pd.to_datetime(player_df['Date'])
    player_df.set_index('Date', inplace=True)
    
    player_df['prev3_avg'] = player_df.H.rolling(3).sum() / player_df.PA.rolling(3).sum()
    player_df['prev5_avg'] = player_df.H.rolling(5).sum() / player_df.PA.rolling(5).sum()
    player_df['next_game_hit'] = (player_df.H.shift(-1) != 0)
    print(player_df.head())
    custom_df.append(player_df)
```
