## Cleaning
First, I will make sure that the season DataFrames are clear of any duplicate observations.

Want to make sure I don't have too many infrequent players.....

```python
fig, ax = plt.subplots(3,1, figsize=(10, 15))
ax[0].hist(daily_2016.playerId.value_counts(), bins=20)
ax[0].set_title('Distribution of 2016 Player Frequency')
ax[0].grid
ax[1].hist(daily_2017.playerId.value_counts(), bins=20)
ax[1].set_title('Distribution of 2017 Player Frequency')
ax[2].hist(daily_2018.playerId.value_counts(), bins=20)
ax[2].set_title('Distribution of 2018 Player Frequency')
[ax[x].grid(alpha=.5) for x in range(3)];
```

That's a lot of players who only appear in a small number of the season's games. If a player plays that infrequently, it seems a reasonable assumption that his skills may not be comparable to that of an everyday player. He may also be a pitcher, particularly in the American League where his batting opportunities would be limited to interleague away games. In any event, the performance of these infrequent batters are unlikely to be useful for predicting the performance of more offensive-focused players, and even the best bench players represent a much riskier proposition as Beat the Streak picks, so being able to predict their performance is not a priority for this project. For these reasons, I will drop all observations for players who appear in less than one quarter of the season's games.

The `remove_under25` function will take in a season's DataFrame and interate through all players in that DataFrame. For each player that has fewer than 41 appearances, the function will overwrite the DataFrame to contain all observations except for that player's. Once all players have been checked, it will return the more exclusive DataFrame.

```python
def remove_under25(df):
    for player in df.playerId.unique():
        if df[df['playerId'] == player].Date.count() < 41:
            df = df[df['playerId'] != player]
    return df
```

Now I will remove the infrequent batters by running the `remove_under25` function on each season and overwriting each season's DataFrame with the function's return.

```python
daily_2016 = remove_under25(daily_2016)
daily_2017 = remove_under25(daily_2017)
daily_2018 = remove_under25(daily_2018)
```

The following function will iterate through each player in the season DataFrame, create an individual timeseries DataFrame for that player's season data, and calculate new features for use in modeling. The timeseries component will allow me to introduce recent trends into each observation, as 3- and 5-game rolling stats will be calculated. Once each player's stats have been calculated, the player's individual DataFrame will be appended back into a collective DataFrame and when every player has been addressed, the function will return the comprehensive DataFrame that has been resorted by date. Because these calculations will take a long time to finish, a counter an print statements will be used to display progress as the function runs.

```python
def add_custom_stats(df):
    custom_df = pd.DataFrame()
    counter = 0
    
    for player in df.playerId.unique():
        player_df = df[df['playerId'] == player]
        player_df['Date'] = pd.to_datetime(player_df['Date'])
        player_df.set_index('Date', inplace=True)
        
        player_df['prev3_pa'] = player_df.PA.rolling(3).sum()
        player_df['prev5_pa'] = player_df.PA.rolling(5).sum()
        
        player_df['prev3_bippa'] = player_df.PA.rolling(3).sum() - player_df.BB.rolling(3).sum() - player_df.SO.rolling(3).sum()
        player_df['prev5_bippa'] = player_df.PA.rolling(5).sum() - player_df.BB.rolling(5).sum() - player_df.SO.rolling(5).sum()
        
        player_df['hpa'] = player_df.H / player_df.PA
        player_df['prev3_hpa'] = player_df.H.rolling(3).sum() / player_df.PA.rolling(3).sum()
        player_df['prev5_hpa'] = player_df.H.rolling(5).sum() / player_df.PA.rolling(5).sum()
        
        player_df['next_game_hit'] = (player_df.H.shift(-1) != 0)
        
        custom_df = custom_df.append(player_df)
        
        counter += 1
        if counter % 25 == 0:
            print(f'{counter} of {df.playerId.nunique()} players complete.')

    print('DataFrame complete.')
        
    return custom_df.sort_index()
```

```python
all_seasons.max()
```

That max value of `inf` in the newly-created `hpa` column could be trouble. I'll take a closer look.

```python
all_seasons.sort_values('hpa', ascending=False).head(3)
```

There only seems to be the one bad instance. Since `hpa` is calculated by dividing `H` by `PA`, I can see that the bad entry has a 0 value for `PA`, which must have caused a "divide by zero" operation. The other stats in that observation show that Phil Gosselin had recorded one home run in a single at-bat during that game. Since it is impossible for a player to record a hit or have an at-bat without also making at least one plate appearance, this is almost certainly a typo. I'll check the `PA` values to confirm that it's an isolated occurrence.

```python
all_seasons.PA.value_counts()
```

As that is the only instance of a 0 value I'm happy to consider it a typo and remove it from the dataset.

```python
all_seasons = all_seasons[all_seasons.PA != 0]
```
