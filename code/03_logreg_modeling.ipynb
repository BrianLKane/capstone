### Notebook Summary
In this notebook I will do my first modeling on the aggregated batter data. My first approach will be a logistic regression classifier. Logistic regression is simple but effective, and allows for greater interpretability than other modeling methods.

### Library and Data Import

```python
import pandas as pd
import numpy as np
import seaborn as sns

import pickle

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
all_seasons = pd.read_csv('../data/batters_all_seasons_custom_dated.csv')
```

### Train Test Split / Scaling

First, I will create my X and y DataFrames. As mentioned in the previous notebook, I am not going to feed the player names, teams, or player IDs into the X dataset both for computational efficiency as well as to ensure a generalized model focused only on player performance. I will also keep the `CS` and `SB` out of the X dataset since base stealing is objectively unrelated to hitting performance.

The `next_game_hit` target will be assigned to y and isolated from X.

```python
X = all_seasons.drop(columns=['Date', 'Name', 'Tm',
                              'playerId', 'CS', 'SB',
                              'next_game_hit'])
y = all_seasons['next_game_hit']
```

Next, I will use `train_test_split` to create my training and testing sets. Although each observation in X should be independent, because of the time component to this data I am setting `shuffle` to False. This also eliminates the need for an assigned random state.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False)
```

I will instantiate and fit a `StandardScaler` to the X training data and apply that transformation to both the training and test X data.

```python
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

### Instantiating & Fitting the Model

I will now instantiate a `LogisticRegression` model with the current default solver of `liblinear` specified. (May 2019)

```python
lr = LogisticRegression(solver='liblinear')
```

And now I will fit the model on my scaled X training data and the y training data and check its accuracy score on the training data and the testing data.

```python
lr.fit(X_train_sc, y_train)
print(f'Training Accuracy: {lr.score(X_train_sc, y_train)}')
print(f'Testing Accuracy: {lr.score(X_test_sc, y_test)}')
```

### Model Evaluation

With all of the classification model scoring that I'm going to be doing in this project, it will be helpful to have easy access to a printed confusion matrix of a given model's performance. I will define a function to do just that in the cell below. In addition to printing a confusion matrix for the model's predictions, the function print the model's accuracy and precision scores, since accuracy is a good general check on the model's performance and precision is going to be the scoring metric for which I optimize all models in this project. For a more thorough discussion on the precision metric as it relates to this project, please refer to the technical report in the project's GitHub repository.

```python
def score_model(actuals, predictions):
    tn, fp, fn, tp = confusion_matrix(actuals, predictions).ravel()
    print(f'T Pos: {tp},   F Pos: {fp}')
    print(f'F Neg: {fn},   T Neg: {tn}')
    print(f'Accuracy: {(tp+tn)/len(predictions)}')
    print(f'Precision: {tp/(tp+fp)}')
```

Now I will use the model to make predictions on the scaled X testing data and store them in `lr_preds`.

```python
lr_preds = lr.predict(X_test_sc)
```

Before running the model's predictions through my scoring function, I will check the mean value of `y_test`. Because this is a binary classification problem, the actual y test values are either `True` or `False` (0 or 1). Therefore the mean of `y_test` will be the same as my accuracy score if I were to guess `True` for all observations. This will serve as a baseline score for both precision and accuracy (since both scores would be equal in the instance of an "all True" prediction) against which to measure model performance.

```python
y_test.mean()
```

Now I will score my model's real predictions against the actual y values.

```python
score_model(y_test, lr_preds)
```

That's a very modest improvement over the baseline, but I am encouraged by the fact that precision is outperforming accuracy. Now I'll generate a DataFrame of the model's predicted probabilities for each of the testing observations to gauge the model's certainty.

```python
logreg_pred_proba = [i[1] for i in lr.predict_proba(X_test_sc)]

logreg_pred_df = pd.DataFrame({'true_values': y_test,
                           'pred_probs':logreg_pred_proba})
```

And now I'll visualize the distribution of those predicted probabilities.

```python
plt.figure(figsize = (16,8))

plt.hist(logreg_pred_df['pred_probs'], bins=25)

plt.title('Distribution of P(Outcome = 1)', fontsize=22)
plt.ylabel('Frequency', fontsize=18)
plt.xlabel('Predicted Probability that Outcome = 1', fontsize=18)
plt.xlim(0,1);
```

This is a little discouraging. The model's predicted probabilities range roughly from .3 to .7, with only a few stretching toward .8, so in other words the model is not particularly confident in its predictions. Both in negative and positive cases, the model cannot strongly commit to a prediction. It's also important to see that the model is predicting few negative outcomes as the left-skewed distribution of probability is bunching up in the .6 range. I'd like to see how well the model is performing given these probabilities, so now I will visualize the model's predicted probabilities broken out by the actual values.

```python
plt.figure(figsize = (16,8))

plt.vlines(x=0.5,
           ymin = 0,
           ymax = 8,
           color='r',
           linestyle = '--')

sns.distplot(logreg_pred_df[logreg_pred_df['true_values'] == 0]['pred_probs'],
             hist=False, kde=True, kde_kws={'shade': True, 'linewidth': 3},
             label='No Hit')
sns.distplot(logreg_pred_df[logreg_pred_df['true_values'] == 1]['pred_probs'],
             hist=False, kde=True, kde_kws={'shade': True, 'linewidth': 3},
             label='Hit')

plt.title('Distribution of Prediction Probabilities', fontsize=25)
plt.ylabel('Density', fontsize=18)
plt.xlabel('Predicted Probability that Outcome = 1', fontsize=18)
plt.xlim(0,1)

plt.legend(fontsize=20);
```

Again, this is not a promising plot. The dashed red line indicates the .5 probability threshold. Anything to the right of that line is a prediction of "hit" from the model and anything to the left of the line indicates a prediction of "no hit," so the mass of blue on the positive side of this plot is discouraging. Though the negative predictions are shifted slightly more toward the negative side of the plot than the positive predictions are, they both have very similar distributions. I'd really like to see more separation between the two actual value distributions.

For now, what I can do is optimize the model for precision, which I will demonstrate in the following visualization.

```python
plt.figure(figsize = (16,8))

plt.vlines(x=0.59,
           ymin = 0,
           ymax = 8,
           color='r',
           linestyle = '--')

sns.distplot(logreg_pred_df[logreg_pred_df['true_values'] == 0]['pred_probs'],
             hist=False, kde=True, kde_kws={'shade': True, 'linewidth': 3},
             label='No Hit')
sns.distplot(logreg_pred_df[logreg_pred_df['true_values'] == 1]['pred_probs'],
             hist=False, kde=True, kde_kws={'shade': True, 'linewidth': 3},
             label='Hit')



plt.title('Distribution of Prediction Probabilities', fontsize=25)
plt.ylabel('Density', fontsize=18)
plt.xlabel('Predicted Probability that Outcome = 1', fontsize=18)
plt.xlim(0,1)

plt.legend(fontsize=20);
```

While the probabilty distributions are the same in this plot, I have shifted the decision threshold (indicated by the dashed red line) from .5 to .59 to visualize different predictions. The model's returned predicted probability for each observation is unchanged, but just as in the previous plot, only everything to the right of the line is predicted as a hit. So if the model predicts that the probabilty of a given observation producing a hit is .55, even though that technically means it's predicted to be more likely than not, it will now predict "no hit." I will generate a new set of predictions with the .59 decision threshold in the following cell.

```python
lr_preds59 = (lr.predict_proba(X_test_sc)[:,1] >= 0.59).astype(bool)
```

And now I will rescore the model.

```python
score_model(y_test, lr_preds59)
```

We can see that increasing the decision threshold has made the model's predictions less accurate, but they are more precise. Since the specific context of Beat the Streak calls for a precision-optimized model, I will gladly accept this accuracy/precision tradeoff. With this model predicting on a decision of .59, we have seen roughly a 5% precision improvement over the baseline model, which had precision and accuracy scores of 0.59.

### Interpreting the Most Influential Features

Given the model's current performance, this may be a moot point. But because interpretability is one of the main attributes of logistic regression, I want to take a moment to examine the model's most predictive features and their weights. The following cell will create a DataFrame of each feature in the X training/testing set corresponding to its predictive weight in the model. Then it will generate a visualization of the 10 most heavily-weighted features and their positive/negative influence.

```python
coef_df = pd.DataFrame({'features': X_test.columns,
                           'weights':lr.coef_[0][:50]})
coef_df.set_index(coef_df.features, inplace=True)
coef_df['abs_w'] = coef_df.weights.abs()
weights = coef_df.sort_values('abs_w', ascending=False)['weights'].head(10)
labels = weights.index

plt.figure(figsize=(20,10))
plt.barh(labels, weights)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Weights', fontsize=30)
plt.ylabel('Features', fontsize=30)
plt.grid(alpha=.5)
plt.title(f'Most Important Features and Weights', fontsize=40);
```

According to this plot, the two features most influential features in making a positive prediction are Weighted On Base Average and Weighted Runs Created. The two most influential features in making a negative prediction are Weighted Runs Above Average and Weighted Runs Created. All of these stats are reasonable candidates to be among the most influential features because (in the case of wRC, wRAA, and wOBA) they reflect a comprehensive approach to measuring offensive performance and wRC+ takes venue and context into account. And wRAA is tied to generating runs as a hitter, so it makes sense that it would have great influence in the model. But it does concern me that wRC+ and wRAA are both exerting their strong influence in a negative direction rather than a positive one since the higher those stats are the more indicative of hitting prowess I would expect them to be.
